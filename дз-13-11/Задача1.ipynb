–°—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞—Ç—å –º–∞—Å—Å–∏–≤ –¥–∞–Ω–Ω—ã—Ö —Å —à–∞–≥–æ–º 0.01 –¥–ª—è —Ñ—É–Ω–∫—Ü–∏–∏ –Ω–∏–∂–µ(—Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –ø–∞–∫–µ—Ç–∞ NumPy)–∏ –æ—Ç–æ–±—Ä–∞–∑–∏—Ç—å —ç—Ç–∏ –¥–∞–Ω–Ω—ã–µ –Ω–∞ —ç–∫—Ä–∞–Ω–µ –≤ –≤–∏–¥–µ –æ–±—ã—á–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞ –≤ –¥–µ–∫–∞—Ä—Ç–æ–≤—ã—Ö –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω—ã—Ö –æ—Å—è—Ö —Å –ø–æ–º–æ—â—å—é –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ matplotlib.
–ì—Ä–∞—Ñ–∏–∫ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –≤—ã–Ω–µ—Å–µ–Ω –≤ –æ—Ç–¥–µ–ª—å–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é. –§—É–Ω–∫—Ü–∏—è:ùë¶(ùë•)=ùëêùëúùë†(20¬∑ùë•)/(ùë•+0,1),ùë•Œµ[0;4]

```python
import numpy as np
import math

x = np.arange(0, 4.1, 0.1)
y = np.array([math.cos(20*i)/(i + 0.1) for i in x])
```

```python
import matplotlib.pyplot as plt

def graphMatplot(x,y):
    plt.figure(figsize=(6, 3)) 
    plt.plot(x, y)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('–ì—Ä–∞—Ñ–∏–∫ Matplotlib')
    plt.show()

graphMatplot(x,y)
```

–û—Ç–æ–±—Ä–∞–∑–∏—Ç—å —Ç–æ—Ç –∂–µ –≥—Ä–∞—Ñ–∏–∫ —Å –ø–æ–º–æ—â—å—é –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ seaborn.–£–±–µ–¥–∏—Ç—å—Å—è,—á—Ç–æ –≤—Å–µ –Ω–∞—Å—Ç—Ä–æ–π–∫–∏ –∑–∞–¥–∞–Ω—ã –≤–µ—Ä–Ω–æ. –ì—Ä–∞—Ñ–∏–∫ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –≤—ã–Ω–µ—Å–µ–Ω –≤ –æ—Ç–¥–µ–ª—å–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é.

```python
import seaborn as sns
import matplotlib.pyplot as plt

def graphSeaborn(x,y):
    plt.figure(figsize=(6, 3)) 
    sns.lineplot(x=x, y=y)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('–ì—Ä–∞—Ñ–∏–∫ Seaborn')
    plt.show()

graphSeaborn(x,y)
```

–û—Ç–æ–±—Ä–∞–∑–∏—Ç—å —Ç–æ—Ç –∂–µ –≥—Ä–∞—Ñ–∏–∫ —Å –ø–æ–º–æ—â—å—é –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ plotly.–£–±–µ–¥–∏—Ç—å—Å—è,—á—Ç–æ –≤—Å–µ –Ω–∞—Å—Ç—Ä–æ–π–∫–∏ –∑–∞–¥–∞–Ω—ã –≤–µ—Ä–Ω–æ. –ì—Ä–∞—Ñ–∏–∫ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –≤—ã–Ω–µ—Å–µ–Ω –≤ –æ—Ç–¥–µ–ª—å–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é.

```python
import plotly.graph_objects as go

def graphPlotly(x,y):
    fig = go.Figure(data=go.Scatter(x=x, y=y, mode='lines'))
    fig.update_layout(title='–ì—Ä–∞—Ñ–∏–∫ Plotly',
                    xaxis_title='X', 
                    yaxis_title='Y', width = 600, height = 350
                    )
    fig.show()

graphPlotly(x,y)
```
