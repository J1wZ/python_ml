1.Сохранить любую цветную фотографию размером 300x500 пиксело в вформате JPEG(jpg).
2.Загрузить в программу и отобразить фотографию в полноцветном варианте(RGB)и в градациях серого(grayscale).Используйте библиотеку matplotlib.

```python
import matplotlib.pyplot as plt
import matplotlib.image as img 

image = img.imread('краски.jpg') 

plt.imshow(image)
plt.show()
```

```python
import matplotlib.pyplot as plt
from PIL import Image
image = Image.open('краски.jpg').convert("L")
plt.imshow(image, cmap='gray')
plt.show()
```
