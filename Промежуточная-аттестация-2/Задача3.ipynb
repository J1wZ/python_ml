```python
from sqlalchemy import create_engine, text
import pandas as pd
    
user = 'root'
password = 'NoMorFort19#'
host = '127.0.0.1'
port = 3306
database = 'dvdrental'

def get_connection():
    return create_engine(
        url="mysql+pymysql://{0}:{1}@{2}:{3}/{4}".format(user, password, host, port, database))

def exec_query(query):
    try:
        engine = get_connection()
        print(f"Успешно подключено.")
        with engine.connect() as connection:
            df = pd.DataFrame(connection.execute(text(query)))
        return df
    except Exception as ex:
        print("Не получилось подключиться: \n", ex)
```

```python
query = """
    SELECT
    film.title as film_title, film.replacement_cost, film.rental_duration, film.rating as film_rating, film.special_features, film.length as film_length,
    CONCAT_WS(' ',actor.first_name, actor.last_name) as actor_name,
    category.name as Category
from film
join film_actor on film.film_id = film_actor.film_id
join actor on actor.actor_id = film_actor.actor_id
join film_category on film_category.film_id = film.film_id
join category on film_category.category_id = category.category_id
order by film_title desc
    """
merged_db = exec_query(query)
print(merged_db)
merged_db.to_csv('merged_db.csv', sep=',', index=False, encoding='utf-8')
```

```python
import pandas as pd

df=pd.read_csv('merged_db.csv')
df.columns = ['Название фильма', 'Цена замены','Продолжитительность аренды','Возрастной рейтинг',
              'Особые харак-ки','Продолжительность фильма','Актер',
              'Категория']
df
```

(a) для каждой числовой переменной вычислить:

```python
numeric_df = df.select_dtypes(include=["number"])
numeric_df
```

```python
#Подсчитываем сколько процентов незаполнено столбце
percent_missing = numeric_df.isnull().sum() * 100 / len(numeric_df.index)
missing_numeric_value_df = pd.DataFrame({'Незаполнено столбце %': percent_missing})

missing_numeric_value_df
```

```python
#Находим максимальные и минимальные значения
min_max_df = pd.DataFrame({'Максимальное значение': numeric_df.max(), 'Минимальное значение': numeric_df.min()})

min_max_df
```

```python
#Находим средние значения
mean_df = pd.DataFrame({'Среднее значение': numeric_df.mean()})

mean_df
```

```python
#Находим медиану
median_df = pd.DataFrame({'Медиана': numeric_df.median()})

median_df
```

```python
#Находим дисперсию
var_df = pd.DataFrame({'Диспрерсия': numeric_df.var()})

var_df
```

```python
#Находим квантиль 0.1 и 0.9
quantile_df = pd.DataFrame({'Квантиль 0.1': numeric_df.quantile(0.1), 'Квантиль 0.9': numeric_df.quantile(0.9)})

quantile_df
```

```python
#Находим квартиль 1 и 3
quartile_df = pd.DataFrame({'Квартиль 1': numeric_df.quantile(0.25), 'Квартиль 3': numeric_df.quantile(0.75)})

quartile_df
```

```python
numeric_df_analysis = pd.concat([missing_numeric_value_df, min_max_df, mean_df, median_df, var_df, quantile_df, quartile_df], axis = 1)
numeric_df_analysis.to_csv('numeric_df_analysis.csv', sep=',', index=False, encoding='utf-8')
numeric_df_analysis
```

(b) для каждой категориальной переменной вычислить:

```python
#Выбираем все столбцы с категориальными значениями
category_df = df[['Возрастной рейтинг',
              'Особые харак-ки',
              'Категория']]
category_df
```

```python
#Подсчитываем сколько процентов незаполнено столбце
percent_missing = category_df.isnull().sum() * 100 / len(category_df.index)
missing_category_value_df = pd.DataFrame({'Незаполнено столбце %': percent_missing})

missing_category_value_df
```

```python
#Находим кол-во уникальных значений
unique_df = pd.DataFrame({'Уникальных значений в столбце': category_df.nunique()})
unique_df
```

```python
#Находим моду столбцов
mode_df =  pd.DataFrame({'Мода в столбце':category_df.mode().loc[0]})
mode_df
```

```python
category_df_analysis = pd.concat([missing_category_value_df, unique_df, mode_df], axis = 1)
category_df_analysis.to_csv('category_df_analysis.csv', sep=',', index=False, encoding='utf-8')
category_df_analysis
```
