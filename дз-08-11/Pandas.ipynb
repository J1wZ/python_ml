Найдите количество уникальных значений столбца «relative humidity (%)»и отобразите их на экране.

Найдите 10 наибольших значений «windspeed(m/s)»и постройте по ним график с помощью встроенных возможностей pandas DataFrame.

```python
top10 = df.loc[:,['date-time','wind speed (m/s)']].nlargest(10, ['wind speed (m/s)'])
top10
```

```python
top10.cumsum()
top10.plot(x='date-time', y='wind speed (m/s)', figsize=(10,5))
```
