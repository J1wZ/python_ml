Загрузим файл JCMB_2015_May.csv по ссылке в DataFrame

```python
import pandas as pd

df=pd.read_csv('https://www.geos.ed.ac.uk/~weather/jcmb_ws/JCMB_2015_May.csv')
df.columns = ['Дата', 'Атмосферное_давление','Осадки','Скорость_ветра',
              'Направление_ветра','Температура','Относительная_влажность_воздуха',
              'Солнечный_поток','Батарея']
df
```

Сохранить первые 1000 строк DataFrame в CSV-файл

```python
df.head(1000).to_csv('May_2015.csv', index=False)
```

Найдите 10 наибольших значений «windspeed(m/s)»и постройте по ним график с помощью встроенных возможностей pandas DataFrame.

```python
top10 = df.loc[:,['Дата','Скорость_ветра']].nlargest(10, ['Скорость_ветра'])
top10
```

```python
top10.cumsum()
top10.plot(x='Дата', y='Скорость_ветра', figsize=(10,5))
```
